```python
import pandas as pd

# Cargar el DataFrame
df_wine = pd.read_csv("./data/winemag-data-130k-v2.csv")

# Observar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df_wine.head())

# Resumen del DataFrame
print("\nResumen del DataFrame:")
print(df_wine.info())

# Estadísticas descriptivas del DataFrame
print("\nEstadísticas descriptivas del DataFrame:")
print(df_wine.describe())

# Valores únicos en cada columna
print("\nValores únicos en cada columna:")
print(df_wine.nunique())

# Conteo de valores en la columna 'country'
print("\nConteo de valores en la columna 'country':")
print(df_wine['country'].value_counts())

# Verificar valores nulos en el DataFrame
print("\nValores nulos en el DataFrame:")
print(df_wine.isnull().sum())

# Renombrar columnas
df_wine.rename(columns={
    'country': 'pais',
    'points': 'puntuacion',
    'price': 'precio',
    'variety': 'variedad_uva'
}, inplace=True)

# Crear nuevas columnas
# 1. Calidad del vino según la puntuación
df_wine['calidad'] = pd.cut(df_wine['puntuacion'], bins=[0, 85, 89, 94, 100], labels=['baja', 'media', 'alta', 'excelente'])

# 2. Rango de precios
df_wine['rango_precio'] = pd.cut(df_wine['precio'], bins=[0, 20, 50, 100, float('inf')], labels=['barato', 'medio', 'caro', 'muy_caro'])

# 3. Longitud del nombre del vino
df_wine['longitud_nombre_vino'] = df_wine['title'].str.len()

# Reporte 1: Promedio de puntuación por país
reporte_promedio_puntuacion = df_wine.groupby('pais')['puntuacion'].mean().sort_values(ascending=False)

# Reporte 2: Precio promedio del vino por país, ordenados de mayor a menor precio promedio
reporte_precio_pais = df_wine.groupby('pais')['precio'].mean().sort_values(ascending=False)

# Reporte 3: Promedio de precio del vino por variedad de uva, ordenados de mayor a menor precio promedio
reporte_precio_variedad = df_wine.groupby('variedad_uva')['precio'].mean().sort_values(ascending=False)

# Reporte 4: Cantidad de vinos por país, ordenados de mayor a menor cantidad de vinos
reporte_cantidad_vinos_pais = df_wine.groupby('pais').size().sort_values(ascending=False)

# Ruta para guardar el archivo CSV
ruta_archivo_csv = "./reporte_cantidad_vinos_pais.csv"

# Guardar el reporte de cantidad de vinos por país en un archivo CSV
reporte_cantidad_vinos_pais.to_csv(ruta_archivo_csv, header=True)

print(f"El reporte de cantidad de vinos por país se ha guardado exitosamente en '{ruta_archivo_csv}'.")


```
